<span style='color:#9c27b0'>**\(Part 1\) Some observations about bipartite graphs**</span>


```python
# Part 1: Bipartite Graphs Adjacency Matrix

# Defined my graph
G = Graph({"a": ["x", "y"],
          "b": ["x", "z"]})

G.is_bipartite()

#

# Make the bipartition
G = BipartiteGraph(G)

# Plot G so that the partitions colored differently
G.plot(partition = [G.left, G.right], figsize = 2)

# Present adjacency matrix
G.adjacency_matrix()

# Reduced form of the matrix
G.reduced_adjacency_matrix()
```

<span style='color:#9c27b0'>**Note: For this reason, any 0\-1 matrix can represent a bipartite graph.**</span>

<span style='color:#9c27b0'>**1, 0, 1**</span>

<span style='color:#9c27b0'>**1, 1, 0**</span>

Take $V_1$ set of rows and $V_2$ set of columns and a vertex representing a row is adjacent to a vertex representing a column if there is a 1 in the corresponding entry.

```python
M = Matrix([[1, 0, 1, 1], [1, 1, 0, 0]])

G = BipartiteGraph(M)

plot(G, partition = [G.left, G.right], figsize = 2)
```

<span style='color:#9c27b0'>**\(Part 3\) A product operation on graphs**</span>


```python
@interact
def f(n_1 = [2..6], n_2 = [3..5]):
    #
    G = graphs.PathGraph(n_1)
    H = graphs.CompleteGraph(n_2)
    K = G.cartesian_product(H)
    #
    print("Graphs G (left), H (center) and K = G*H (right):")
    G.show(vertex_size = 35, vertex_labels = False, figsize = 0.5*n_1)
    H.show(vertex_size = 35, vertex_labels = False, figsize = 0.5*n_2)
    K.show(vertex_size = 35, vertex_labels = False, figsize = 0.3*n_1*n_2)
    #
    print(f"Number of vertices: n = {len(K.vertices())}")
    print(f"Number of edges: m = {len(K.edges())}")
```

<span style='color:#9c27b0'>**\(Part 4\) Experimenting with the complement of bipartite graphs**</span>


```python
@interact
def f(n_1 = [2..7], n_2 = [3..5]):
    #
    G = graphs.RandomBipartite(n_1, n_2, 0.65)
    H = G.complement()
    #
    print(f"Here is the the bipartite graph G on {n_1} and {n_2} vertices in each partition, and its complement H.\n")
    G.show(figsize = min_symbolic([4, 0.4*n_1*n_2]), vertex_labels = False)
    H.show(figsize = min_symbolic([4, 0.4*n_1*n_2]), vertex_labels = False)
    #
    if H.is_bipartite():
        print("H is bipartite.\n")
    else:
        print("H is not bipartite.\n")
```
