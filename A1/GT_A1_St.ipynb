```python
# Run to see the output.

# Defining the edges of my graph; "vertex name": {"neighbour": "edge label"}

G = Graph({"Home": {"Schl": "B", "gym": "BK", "Mom": "W", "F1":"W", "F2": "B", "Wrk": "B", "Mall": "B"},
           "Mall": {"F1": "B", "F2": "B"},
           "Schl": {"gym": "B", "Mom": "B"},
           "Mom": {"Wrk": "B", "Mall": "B"},
           "Wrk": {"F1": "B", "F2": "B", "gym": "B"},
           "F1": {"F2": "B", "Mall": "B"},
           "F2": {"Mall": "W"}})

# Plotting my graph by styling the edges and vertices the way I want:

G.show(vertex_colors = {"red": ["F1", "F2", "Mom"], "pink": ["gym"]},
       edge_thickness = 2.5,
       edge_colors = {"gray": [(a, b) for a, b, c in G.edges() if G.edge_label(a, b) == "W"]},
       edge_labels = True, vertex_size = 800,
       layout = "circular")

# Checking connectivity:

G.is_connected()

# Choosing only edges that are bus-able & plotting it:

BusSubgraph = Graph([(a, b, c) for a, b, c in G.edges() if G.edge_label(a, b) == "B"])

print("Subgraph with all the routes that I can bus:")

plot(BusSubgraph, vertex_size = 500, edge_labels = True, figsize = 4)

# Deciding if I can take busses in a loop and visit all locations:

print("Can I take bus and visit all the locations in a circular form and go back home?")

plot(BusSubgraph.hamiltonian_cycle(), vertex_size = 500, edge_labels = True, figsize = 4)
```
